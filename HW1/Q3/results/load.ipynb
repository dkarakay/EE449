```python
from utils.utils import part3Plots

import json
models=['mlp_1','mlp_2']#Models that will be printed
results=list()
for model in models:
    if(model=='cnn_4'):
        f = open('DUMMY_q2_'+model+'.json')

        # returns JSON object as
        # a dictionary
        results.append(json.load(f))
        f.close()
    else:
        f = open('q2_'+model+'.json')

        # returns JSON object as
        # a dictionary
        results.append(json.load(f))
        f.close()
part3Plots(results, save_dir=r'resultQ2', filename='part2Plots___________')
```
